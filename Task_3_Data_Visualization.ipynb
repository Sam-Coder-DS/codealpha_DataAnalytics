### Transform Raw Data Into Visual Formats Like Charts, Graphs and Dashboards ðŸ‘‡

```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset
df = pd.read_csv("/content/reviews.csv")
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['month'] = df['date'].dt.to_period('M')

# Step 3: Distribution of Ratings per Platform
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', hue='platform', palette='Set2')
plt.title('Rating Distribution per Platform')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.legend(title='Platform')
plt.tight_layout()
plt.show()

# Step 4: Average Rating by Platform
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='platform', y='rating', hue='platform', palette='pastel', legend=False)

plt.title('Average Rating by Platform')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

# Step 5: Reviews Over Time
df['month'] = df['date'].dt.to_period('M').astype(str)

# Reviews Over Time Chart
monthly_reviews = df.groupby(['month', 'platform']).size().reset_index(name='review_count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_reviews, x='month', y='review_count', hue='platform', marker='o')
plt.title('Number of Reviews Over Time')
plt.xlabel('Month')
plt.ylabel('Review Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(review for review in df['review'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Reviews')
plt.show()
```

### Use Tools Like Matplotlib, Seaborn For Creating Visuals ðŸ‘‡

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/reviews.csv")
avg_rating = df.groupby('platform')['rating'].mean()

# Plot using Matplotlib
plt.figure(figsize=(8, 5))
avg_rating.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average Rating per Platform')
plt.ylabel('Average Rating')
plt.xlabel('Platform')
plt.ylim(0, 5)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot using Seaborn
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', hue='platform', palette='Set2')
plt.title('Rating Distribution by Platform')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.legend(title='Platform')
plt.tight_layout()
plt.show()
```

### Design Visuals That Enhance Understanding And Reveals Insights Clearly ðŸ‘‡

```python
avg_rating = df.groupby('platform')['rating'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_rating, x='platform', y='rating', hue='platform', palette='coolwarm', legend=False)
plt.title('Average Rating per Platform')
plt.ylim(0, 5)
plt.ylabel('Average Rating')
plt.xlabel('Platform')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', hue='platform', palette='Set2')
plt.title('Rating Distribution by Platform')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.legend(title='Platform')
plt.tight_layout()
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')  # Convert date column properly
df['month'] = df['date'].dt.to_period('M').astype(str)    # Create month column

monthly_reviews = df.groupby(['month', 'platform']).size().reset_index(name='review_count')
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_reviews, x='month', y='review_count', hue='platform', marker='o')
plt.title('Reviews Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
bubble_data = df.groupby('platform').agg({'rating': ['count', 'mean']}).reset_index()
bubble_data.columns = ['platform', 'review_count', 'avg_rating']

plt.figure(figsize=(10, 6))
plt.scatter(bubble_data['platform'], bubble_data['review_count'],
            s=bubble_data['avg_rating'] * 500, color='orange',
            alpha=0.6, edgecolors='black')

for i in range(len(bubble_data)):
    plt.text(bubble_data['platform'][i], bubble_data['review_count'][i] + 5,
             f"{bubble_data['avg_rating'][i]:.2f}", ha='center')

plt.title("Bubble Plot: Review Count vs Avg Rating")
plt.tight_layout()
plt.show()
```

```python
text = " ".join(df['review'].dropna().astype(str))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Common Words in User Reviews")
plt.show()
```
